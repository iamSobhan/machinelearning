```python
#example plot for the relu activation function
from matplotlib import pyplot

#rectified linear function
def rectified(x):
    return max(0.0, x)

#define input data
inputs = [x for x in range(-20, 20)]

#calculate outputs
outputs = [rectified(x) for x in inputs]

#plot input vs outputs
pyplot.plot(inputs, outputs)
pyplot.show()
```

```python
#example plot for the sigmoid activation function
from math import exp
from matplotlib import pyplot


#sigmoid activation function
def sigmoid(x):
    return 1.0 / (1.0 + exp(-x))

#define input data
inputs = [x for x in range(-10, 10)]

#calculate outputs
outputs = [sigmoid(x) for x in inputs]

#plot input vs outputs
pyplot.plot(inputs, outputs)
pyplot.show()
```

```python
#example plot for the tanh activation function
from matplotlib import pyplot
from math import exp

#tanh activation function
def tanh(x):
    return (exp(x) - exp(-x)) / (exp(x) + exp(-x))

#define input data
inputs = [x for x in range(-10, 10)]

#calculate outputs
outputs = [tanh(x) for x in inputs]

#plot input vs outputs
pyplot.plot(inputs, outputs)
pyplot.show()
```

```python
#example plot for the linear activation function
from matplotlib import pyplot

#linear activation function
def linear(x):
    return x

#define input data
inputs = [x for x in range(-10, 10)]

#calculate outputs
outputs = [linear(x) for x in inputs]

#plot input vs outputs
pyplot.plot(inputs, outputs)
pyplot.show()
```

```python
from numpy import exp

#softmax activation function
def softmax(x):
    return exp(x) / exp(x).sum()

#define input data
inputs = [1.0, 3.0, 2.0]

#calculate outputs
outputs = softmax(inputs)

#report the probabilities
print(outputs)

#report the sum of probabilities
print(outputs.sum())
```
