```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#This Dataset is very Unbalanced. So separating data for further analysis.

#0 --- Normal Transaction
#1 --- Fraud Transaction
```

```python
normal = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
#Under-Sampling
#Build a sample dataset containing similar distribution of normal transactions and fraud transactions.
```

```python
normal_sample = normal.sample(n = 492)
```

```python
#Concatenation of two DataFrames

new_credit_card_data = pd.concat([normal_sample, fraud], axis = 0) #axis = 0 means the dataset added one by one, all the 492 values wiil be added below the normal_sample.

```

```python
new_credit_card_data.groupby("Class").mean()
```

```python
X = new_credit_card_data.drop(columns = "Class", axis = 1)
Y = new_credit_card_data["Class"]
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 1)
```

```python
print(X.shape)
```

```python
print(X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy on Training Data: ", training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy on Test Data: ", test_data_accuracy)
```
